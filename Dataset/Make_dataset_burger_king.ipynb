```python
import pandas as pd
import numpy as np
import random
```

```python
# 1. Konfigurasi Awal
N = 48620
output_path = "burger_king_data_simple.xlsx"

# Data Referensi Menu (Langsung dalam DataFrame agar mudah di-sample)
menu_data = [
    ("whopper", "Whopper Burger", "Beef Patty, Lettuce, Tomato, Onion, Pickles, Mayo"),
    ("bk_chicken", "BK Chicken Burger", "Crispy Chicken, Lettuce, Mayo"),
    ("cheeseburger", "Cheeseburger", "Beef Patty, Cheese, Pickles, Ketchup, Mustard"),
    ("double_steakhouse", "Double Steakhouse Burger", "Double Beef Patty, Crispy Onion, BBQ Sauce"),
    ("plant_based_whopper", "Plant Based Whopper", "Plant Patty, Lettuce, Tomato, Vegan Mayo")
]
df_menu = pd.DataFrame(menu_data, columns=["code", "burger_name", "burger_ingredients"])
```

```python
# 2. Generate Data Utama (Random Sampling sekaligus untuk N baris)
# Kita ambil N baris acak dari menu_data
df = df_menu.sample(n=N, replace=True).reset_index(drop=True)

# 3. Generate Kolom Tambahan secara Vektorisasi (Tanpa Loop)
df['burger_id'] = range(1, N + 1)
df['order_id'] = range(1, N + 1)
df['quantity'] = np.random.randint(1, 4, N)
df['unit_price'] = np.random.choice([25000, 30000, 35000, 40000, 45000], N)
df['total_price'] = df['quantity'] * df['unit_price']
df['burger_size'] = np.random.choice(["S", "M", "L"], N)
df['burger_category'] = np.random.choice(["Beef", "Chicken", "Veggie", "Premium"], N)

# Menggabungkan Code + Size (Contoh: whopper_M)
df['burger_name_id'] = df['code'] + "_" + df['burger_size'].str.lower()
```

```python
# 4. Generate Tanggal & Waktu (Rentang 1 Tahun)
start_date = np.datetime64('2025-01-01')
# Tambah hari acak (0-364) ke start_date
random_days = np.random.randint(0, 365, N)
df['order_date'] = start_date + random_days.astype('timedelta64[D]')

# Generate Jam Acak (String HH:MM:SS)
hours = np.random.randint(9, 23, N) # Jam 9 sampai 22
mins = np.random.randint(0, 60, N)
secs = np.random.randint(0, 60, N)
# Format waktu dengan map string formatting sederhana
df['order_time'] = [f"{h:02d}:{m:02d}:{s:02d}" for h, m, s in zip(hours, mins, secs)]
```

```python
# 5. Rapikan Urutan Kolom & Export
cols_order = [
    'burger_id','order_id','burger_name_id','quantity','order_date','order_time',
    'unit_price','total_price','burger_size','burger_category','burger_ingredients','burger_name'
]
df = df[cols_order]

df.to_excel(output_path, index=False)
print(f"Selesai! Data disimpan di {output_path}")
```
